# Predictive Modeling with Classical Machine Learning

## Iris Flower Classification using Logistic Regression

### Objective
To classify Iris flowers into Setosa, Versicolor, and Virginica species using Machine Learning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)
```

```python
iris = load_iris()

print("Dataset Loaded Successfully")
print("Number of Samples:", len(iris.data))
print("Number of Features:", len(iris.feature_names))
print("Target Names:", iris.target_names)
```

```python
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["species"] = iris.target

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["species"].value_counts().plot(
    kind="bar"
)

plt.title("Iris Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
```

```python
X = iris.data
y = iris.target

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

```python
model = LogisticRegression(max_iter=200)

model.fit(X_train, y_train)

print("Model Trained Successfully")
```

```python
y_pred = model.predict(X_test)

print(y_pred)
```

```python
accuracy = accuracy_score(
    y_test,
    y_pred
)

print("Accuracy:", accuracy)
```

```python
precision = precision_score(
    y_test,
    y_pred,
    average="weighted"
)

print("Precision:", precision)
```

```python
recall = recall_score(
    y_test,
    y_pred,
    average="weighted"
)

print("Recall:", recall)
```

```python
f1 = f1_score(
    y_test,
    y_pred,
    average="weighted"
)

print("F1 Score:", f1)
```

```python
print(
    classification_report(
        y_test,
        y_pred
    )
)
```

```python
cm = confusion_matrix(
    y_test,
    y_pred
)

print(cm)
```

```python
sample = [[5.1, 3.5, 1.4, 0.2]]

prediction = model.predict(sample)

print(prediction)
```

```python
species_names = {
    0: "Iris Setosa",
    1: "Iris Versicolor",
    2: "Iris Virginica"
}

flower = species_names[prediction[0]]

print("Predicted Flower:", flower)
```

# Conclusion

A Logistic Regression model was developed using the Iris Dataset.

The dataset was preprocessed and divided into Training (70%) and Testing (30%) sets.

The model successfully classified Iris flowers into three species:
1. Iris Setosa
2. Iris Versicolor
3. Iris Virginica

The model achieved high Accuracy, Precision, Recall, and F1-Score, demonstrating excellent performance for flower classification.
